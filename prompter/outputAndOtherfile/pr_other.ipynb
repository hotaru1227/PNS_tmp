```python
# 处理一下cpm test数据集的gt成上面那个格式
import json
import numpy as np
import re
dataset_list = np.load("/data/hotaru/projects/PNS_tmp/segmentor/datasets/cpm17_test_files.npy")
# 读取JSON文件
with open('/data/hotaru/projects/PNS_tmp/prompter/datasets/cpm17/test.json', 'r') as f:
    json_data = json.load(f)
png_list = []
for mat_file in dataset_list:
    png_list.append(mat_file.split(".")[0]+".png")
# 提取需要的字段

pattern = r'([^\/\\]+)\.png$'
desired_fields = png_list  # 假设你需要提取的字段名称列表
img_path = '/data/hotaru/projects/PNS_tmp/segmentor/prompts/cpm17/'
extracted_data = {field: json_data[field] for field in desired_fields}
for i in png_list:
    image_name = re.search(pattern, i)[0].split('.')[0]
    # print(re.search(pattern, i)[0].split('.')[0])
    points = []
    for j in extracted_data[i][0]:
        j.append(0)
        # print(j)
        points.append(j)
    # print(points)
    # print(len(extracted_data[i][0]))    
    np.save(img_path+image_name+'.npy', points)
    # break
# print(extracted_data['datasets/cpm17/test/Images/image_01.png'][0])
# 将提取的数据保存为npy文件
# np.save('extracted_data.npy', extracted_data)
```
